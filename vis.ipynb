```python
import numpy as np
import scipy
import sklearn
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
bat, sr = librosa.core.load('./data/ng/bat_chirp.wav')
print(f'Sample Rate: {sr}')
print(f'Duration: {len(bat) / sr} seconds')
spec = librosa.amplitude_to_db(np.abs(librosa.stft(bat)), ref=np.max)
librosa.display.specshow(spec)
```
